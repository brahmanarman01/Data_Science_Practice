```python
# Import the pandas library, which is essential for data manipulation and analysis in Python.
import pandas as pd
```

```python
# Creating DataFrames
# This section demonstrates different methods for creating Pandas DataFrames, a fundamental data structure in data science used for organizing and analyzing tabular data.
```

```python
# Create a DataFrame from a Python list of lists.
# This is a common way to create a DataFrame when you have data structured as rows and columns in a list format.

# Define a list of lists where each inner list represents a row of data.
data = [
    ["Alice", 25],
    ["Bob", 30],
    ["Charlie", 35]
]

# Create a Pandas DataFrame from the list of lists.
# The 'columns' argument is used to specify the names of the columns in the DataFrame.
df = pd.DataFrame(data, columns=["Name", "Age"])

# Display the resulting DataFrame.
# This shows the structure and content of the DataFrame created from the list.
df
```

```python
# from dict
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}

df = pd.DataFrame(data)
df
```

```python
# from numpy array
import numpy as np

arr = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(arr, columns=["A", "B"])
df
```
