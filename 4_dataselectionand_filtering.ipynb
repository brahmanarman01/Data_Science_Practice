```python
# Select the row at index 1 using iloc (integer-location based indexing)
df.iloc[1]
```

```python
# Select rows from index 0 to 2 (inclusive) and specific columns "Actor", "Film", and "Year" using loc (label-based indexing)
df.loc[0:2,["Actor","Film"      ,"Year"]]
```

```python
# Select rows from index 0 to 2 (exclusive of 2) and columns from index 0 to 2 (exclusive of 2) using iloc
df.iloc[0:2,0:2]
```

```python
# Select the value at row index 7 and column label 'Actor' using at (for fast label-based scalar access)
df.at[7,'Actor']
```

```python
# Select the value at row index 0 and column index 1 using iat (for fast integer-location based scalar access)
df.iat[0,1]
```

```python
df[df['IMDb']>7]
```

```python
df[(df["IMDb"] > 7) & (df["Genre"] == "Action")]
```

```python
df[(df["IMDb"] > 7) | (df["Genre"] == "Action")]
```

```python
# The .query() method in pandas lets you filter DataFrame rows using a string expression  it's a more readable and often more concise alternative to using boolean indexing.
```

```python
df.query("IMDb>6 and Genre=='Action'")
```

```python
col = "IMDb"
df.query(f"{col} > 6")
```
