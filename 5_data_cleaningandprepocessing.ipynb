```python
import pandas as pd
# Imports the pandas library, which is used for data manipulation and analysis.
```

```python
df=pd.read_csv('/content/1745501008647-data_cleaning_sample.csv')
# Reads the CSV file into a pandas DataFrame named 'df'.
```

```python
df
# Displays the entire DataFrame 'df'.
```

```python
df.isnull()
# Checks for null (missing) values in each cell of the DataFrame and returns a boolean DataFrame of the same shape.
```

```python
df.isnull().sum()
# Calculates and displays the sum of null values for each column in the DataFrame.
```

```python
df2=df
# Creates a copy of the DataFrame 'df' and assigns it to 'df2'.
```

```python
df2
# Displays the entire DataFrame 'df2'.
```

```python
df2.dropna()
# Drops rows with any null values from 'df2' and displays the result (does not modify 'df2' in place).
```

```python
df2.dropna(axis=1)
# Drops columns with any null values from 'df2' and displays the result (does not modify 'df2' in place).
```

```python
df2.fillna(0)
# Fills all null values in 'df2' with 0 and displays the result (does not modify 'df2' in place).
```

```python
df2["Age"].fillna(df2["Age"].mean())
# Fills null values in the 'Age' column of 'df2' with the mean of the 'Age' column and displays the result (does not modify 'df2' in place).
```

```python
df2.ffill()
# Fills null values using the forward fill method (propagates the last valid observation forward) and displays the result (does not modify 'df2' in place).
```

```python
df2.bfill()
# Fills null values using the backward fill method (propagates the next valid observation backward) and displays the result (does not modify 'df2' in place).
```

```python
df2.duplicated()
# Checks for duplicate rows in 'df2' and returns a boolean Series.
```

```python
df2.drop_duplicates()
# Drops duplicate rows from 'df2' and displays the result (does not modify 'df2' in place).
```

```python
df.duplicated(subset=["Name", "Age"])
# Checks for duplicate rows based on the 'Name' and 'Age' columns in 'df' and returns a boolean Series.
```

```python
df["City"].str.contains("delhi", case=False)
# Checks if the 'City' column contains the string "delhi" (case-insensitive) and returns a boolean Series.
```

```python
df["Email"].str.split("@")
# Splits the strings in the 'Email' column by the "@" symbol and returns a list of strings for each row.
```

```python
# We can always chain methods like .str.strip().str.upper() for clean-up.
# This is a comment indicating that string methods can be chained for data cleaning.
```

```python
df2['Age'].dtypes
# Displays the data type of the 'Age' column in 'df2'.
```

```python
df["Age Group"] = df["Age"].apply(lambda x: "Adult" if x >= 18 else "Minor")
# Creates a new column 'Age Group' based on the 'Age' column, categorizing individuals as "Adult" or "Minor".
df["Age Group"]
# Displays the 'Age Group' column.
```

```python
gender_map = {"M": "Male", "F": "Female"}
# Creates a dictionary to map gender abbreviations to full names.
df["Gender"] = df["Gender"].map(gender_map)
# Maps the values in the 'Gender' column using the 'gender_map' dictionary.
```

```python
df[
    'Gender'
]
# Displays the 'Gender' column after mapping the values.
```

```python
df["City"].replace({"Del": "Delhi", "Mum": "Mumbai"})
# Replaces "Del" with "Delhi" and "Mum" with "Mumbai" in the 'City' column (does not modify the column in place).
df["City"]
# Displays the 'City' column.
```
