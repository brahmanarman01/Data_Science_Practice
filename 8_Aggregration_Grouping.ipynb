```python

import pandas as pd
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Team")["Salary"].mean()     # Calculate the average salary per team
```

```python
df.groupby("Team")["Salary"].sum()      # Calculate the total salary per team
```

```python
df.groupby("Team")["Salary"].count()    # Count the number of employees per team
```

```python
df.groupby("Team")["Salary"].min()      # Find the minimum salary per team
```

```python
df.groupby("Team")["Salary"].max()      # Find the maximum salary per team
```

```python
df.groupby(["Team", "Gender"])["Salary"].mean() # Calculate the average salary per team and gender
```

```python
df.groupby("Team")["Salary"].agg(["mean", "max", "min"]) # Calculate multiple aggregations (mean, max, min) for salary per team
```

```python
df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean") # Add a new column with the average salary for each employee's team
```

```python
df["Team Avg"] # Display the new 'Team Avg' column
```

```python
df # Display the DataFrame with the new 'Team Avg' column
```

```python
df.groupby("Team").filter(lambda x: x["Salary"].mean() > 80) # Filter teams where the average salary is greater than 80
```
